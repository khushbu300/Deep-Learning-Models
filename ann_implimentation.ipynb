```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/aps_data.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df = df.replace('na',np.nan)
```

```python
pd.options.display.max_rows = 1000
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
pd.DataFrame(df.isnull().sum().sort_values(ascending = False)).rename(columns={0:'Null Value Count'})
```

```python
for i in [i for i in df.columns if i not in ['class']]:
  df[i] = df[i].fillna(df[i].median())
```

```python
df.isnull().sum().sum()
```

```python
df.head()
```

```python
x = df.drop('class', axis = 1)
```

```python
y = df['class']
```

```python
y
```

```python
x.info(verbose = True, show_counts = True)
```

```python
y = pd.get_dummies(y, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
x_train = sc.fit_transform(X_train)
x_test = sc.transform(X_test)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()
classifier.add(Dense(units=10, kernel_initializer='he_uniform',activation='relu',input_dim=170))

classifier.add(Dense(units = 10, kernel_initializer = 'he_uniform', activation = 'relu'))

classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation= 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.summary()
```

```python
model_history = classifier.fit(x_train, y_train, batch_size = 10, epochs = 30, validation_split = 0.25)
```

```python
model_history.history.keys()
```

```python
model_history.history['accuracy']
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model_accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred = y_pred>0.05
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
import time
import os

def saveModel_path(model_dir="SAVED_MODELS"):
    os.makedirs(model_dir, exist_ok=True)
    fileName = time.strftime("Model_%Y_%m_%d_%H_%M_%S_.h5")    
    model_path = os.path.join(model_dir, fileName)
    print(f"your model will be saved at the following location\n{model_path}")
    return model_path
```

```python
classifier.save(saveModel_path())
```

```python
from kerastuner.tuners import RandomSearch
```

```python
def build_model(hp):
  model=Sequential()
  model.add(Dense(units=32,kernel_initializer = 'he_uniform',activation='relu',input_dim=170))
  model.add(Dense(units = 32, kernel_initializer = 'he_uniform',activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  optimizer=hp.Choice("optimizer",values=["adam","sgd","rmsprop","adadelta"])
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner=RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='project1',project_name='aps-fault')
```

```python
tuner.search(x_train,y_train,validation_data=(x_test,y_test),epochs=5)
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
mymodel=tuner.get_best_models(num_models=1)[0]
```

```python
from tensorflow.keras import optimizers
def build_model(hp):
    model = Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(Dense(units=hp.Int('units_' + str(i),
                                            min_value=32,
                                            max_value=512,
                                            step=32),
                               activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
tuner_second=RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='project3',project_name='aps-fault-sensor')
```

```python
tuner_second.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```
